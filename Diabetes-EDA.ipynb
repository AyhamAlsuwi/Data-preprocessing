```python
# 'Glucose Before fasting' is a wrong name for the records, the records indicates
df = df.rename(columns={'Glucose Before fasting':'Glucose after fasting','percentage of occuring diabetes':'Percentage of occuring diabetes'})
```

```python
df
```

### Handling missing values

```python
df.isnull().sum()
```

```python
df['Glucose after fasting'] = df['Glucose after fasting'].fillna(df['Glucose after fasting'].median())
```

```python
df['Glucose Anytime'] = df['Glucose Anytime'].fillna(df['Glucose Anytime'].median())
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df['Sex'] = df['Sex'].fillna(df['Sex'].mode()[0])
```

```python
df['Blood Pressure'] = df['Blood Pressure'].fillna(df['Blood Pressure'].mode()[0])
```

```python
df['Family member with Diabetes past or present'] =df['\
Family member with Diabetes past or present'].fillna(df['Family member with Diabetes past or present'].mode()[0])
```

```python
df['BMI'] = df['BMI'].fillna(df['BMI'].median())
```

```python
df['Pregnancies'] = df['Pregnancies'].fillna(df['Pregnancies'].mode()[0])
```

```python
df.isnull().sum()
```

```python
# pregnant male!! must fix
df['Sex'][df['Pregnancies'] == 1].value_counts()
```

```python
df.loc[df['Sex']=='Male' ,'Pregnancies'] = 0
```

```python
df['Sex'][df['Pregnancies'] == 1].value_counts()
```

### Outliers detection

```python
#no outliers
plt.figure(figsize = (12,6))
plt.subplot(1,4,1)
sns.boxplot(y = df['Glucose after fasting'])
plt.subplot(1,4,2)
sns.boxplot(y = df['Glucose Anytime'])
plt.subplot(1,4,3)
sns.boxenplot(y=df['Age'])
plt.subplot(1,4,4)
sns.boxplot(y = df['BMI'])
plt.tight_layout();
```

```python
''' Percentage of diabetes is increasing with Glucose record after fasting and there are three stages: below 100 ,
between 100 and 130, above 130 glucose record.'''
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.scatterplot(x = df['Glucose after fasting'] , y = df['Percentage of occuring diabetes'])
plt.title('Percentage of diabetes among glucose record after fasting')

plt.subplot(1,2,2)
sns.lineplot(x = df['Glucose after fasting'] , y = df['Percentage of occuring diabetes'])
plt.ylabel('Mean of percentage of occuring diabetes')
plt.title('Percentage of diabetes among glucose record after fasting')

plt.tight_layout();
```

```python
''' Percentage of diabetes is increasing with Glucose record anytime and there are three stages: below 145 ,
between 145 and 205, above 205 glucose record.'''
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
sns.scatterplot(x = df['Glucose Anytime'] , y = df['Percentage of occuring diabetes'])
plt.title('Percentage of diabetes among glucose record anytime')

plt.subplot(1,2,2)
sns.lineplot(x = df['Glucose Anytime'] , y = df['Percentage of occuring diabetes'])
plt.ylabel('Mean of percentage of occuring diabetes')
plt.title('Percentage of diabetes among glucose record anytime')

plt.tight_layout();
```

```python
# as age goes up percentage of diabetes increases, after 40 years old the the percentage is significantly increasing
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
sns.scatterplot(x = df['Age'] , y = df['Percentage of occuring diabetes'])
plt.title('Percentage of diabetes among age')

plt.subplot(1,2,2)
sns.lineplot(x = df['Age'] , y = df['Percentage of occuring diabetes'])
plt.ylabel('Mean of percentage of occuring diabetes')
plt.title('Percentage of diabetes among age')

plt.tight_layout();
```

```python
# females are more likely to to have diabetes
plt.figure(figsize = (12,5))
plt.subplot(1,2,1)
sns.histplot(x = df['Percentage of occuring diabetes'] , hue = df['Sex'] ,bins = 20)
plt.subplot(1,2,2)
df.groupby(df['Sex'])['Percentage of occuring diabetes'].mean().plot(kind = 'bar')
plt.ylabel('Mean of Percentage of occuring diabetes');
```

```python
# peaple with blood pressure has higher percentage of occuring diabetes
plt.figure(figsize = (12,5))
plt.subplot(1,2,1)
sns.histplot(x = df['Percentage of occuring diabetes'] , hue = df['Blood Pressure'],bins = 20)
plt.subplot(1,2,2)
df.groupby(df['Blood Pressure'])['Percentage of occuring diabetes'].mean().plot(kind = 'bar')
plt.ylabel('Mean of Percentage of occuring diabetes');
```

```python
# peaple who have a relative with diabetes has higher percentage of occuring diabetes
plt.figure(figsize = (12,5))
plt.subplot(1,2,1)
sns.histplot(x = df['Percentage of occuring diabetes'] , hue = df['Family member with Diabetes past or present'],bins = 20)
plt.subplot(1,2,2)
df.groupby(df['Family member with Diabetes past or present'])['Percentage of occuring diabetes'].mean().plot(kind = 'bar')
plt.ylabel('Mean of Percentage of occuring diabetes');
```

```python
# Pregnant females has higher percentage of occuring diabetes
plt.figure(figsize = (12,5))
plt.subplot(1,2,1)
sns.histplot(x = df['Percentage of occuring diabetes'] , hue = df.loc[df['Sex'] == 'Female' , 'Pregnancies'],bins = 20)
plt.subplot(1,2,2)
df.groupby(['Sex','Pregnancies'])['Percentage of occuring diabetes'].mean().plot(kind = 'bar')
plt.ylabel('Mean of Percentage of occuring diabetes');
```

```python
# as BMI goes up percentage of diabetes increases.
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
sns.scatterplot(x = df['BMI'] , y = df['Percentage of occuring diabetes'])
plt.title('Percentage of diabetes among age')

plt.subplot(1,2,2)
sns.lineplot(x = df['BMI'] , y = df['Percentage of occuring diabetes'])
plt.ylabel('Mean of percentage of occuring diabetes')
plt.title('Percentage of diabetes among age')

plt.tight_layout();
```
