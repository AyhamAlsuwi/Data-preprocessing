```python
#Target feature distribution
sns.histplot(x = df['Performance Index'],kde = True)
plt.xlabel('Total performance');
```

```python
corr_matrix = df.corr(numeric_only=True)
fig , ax = plt.subplots(figsize = (10,8))
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',ax = ax)
plt.title('correlation matrix');
```

```python
# no unacceptable outliers for Hours Studied
#the plot confirms that  hours studied has strong positive effect on the target with linear trend
#the lowest mark in 9 hours class is better from others (lowest)

sns.boxplot(x=df['Hours Studied'],y= df['Performance Index'])
plt.ylabel('total performance');

```

```python
# no unacceptable outliers for Hours Studied
#the plot shows a nearly perfect linear positive trend
# Students with high previous scores will be of the highest total performance
fig, ax = plt.subplots(1,1,figsize=(10,8))
sns.scatterplot(x=df['Previous Scores'],y= df['Performance Index'],ax = ax);
```

```python
#Extracurricular Activities does not make obvious effect on total performance
sns.histplot(x = df['Performance Index'],hue = df['Extracurricular Activities'],bins=20);
```

```python
# no unacceptable outliers
# geting more sleep hours can help studens but not alot
# but gitting more sleep hours alone does not means always improvement in the target (lazy)
sns.boxplot(x = df['Sleep Hours'],y = df['Performance Index'],data=df);
```

```python
#  solving past papers  can give a slight imporvement total Performance but it does not guaranty good performance as it showed

sns.boxplot(x = df['Sample Question Papers Practiced'],y = df['Performance Index']);
```
