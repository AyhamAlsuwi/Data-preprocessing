## Checking null values.  
** the Cabin feature will not fill it because more than 75% of it is null, it will not give a valuable insites.**

```python
df.isnull().sum()
```

```python
plt.subplot(1,2,1)
sns.histplot(x = df['Age'])
plt.subplot(1,2,2)
sns.histplot(x = df['Embarked'])
plt.tight_layout();
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.isnull().sum()
```

```python
plt.subplot(1,2,1)
sns.histplot(x = df['Age'],kde=True)
plt.subplot(1,2,2)
sns.histplot(x = df['Embarked'])
plt.tight_layout();
```

## Outliers handling

```python
# Age outliers are acceptable

q1 , q3 = df['Age'].quantile([0.25 , 0.75])
iqr_Age = q3 - q1
upperB_Age = q3 + 1.5 * iqr_Age
lowerB_Age = q1 - 1.5 * iqr_Age
plt.axvline(upperB_Age)
plt.axvline(lowerB_Age)
sns.histplot(x = df['Age'])
plt.title('Age distribution showing outliers');
```

```python
sns.histplot(x = df['Fare'],kde=True);
```

```python
# Fare outliers are acceptable (representing high class passengers)

plt.figure(figsize = (12,5))

plt.subplot(1,2,1)
q1 , q3 = df['Fare'].quantile([0.25 , 0.75])
iqr_Fare = q3 - q1
upperB_Fare = q3 + 1.5 * iqr_Fare
lowerB_Fare = q1 - 1.5 * iqr_Fare
sns.histplot(x = df['Fare'])
plt.axvline(upperB_Fare)
plt.axvline(lowerB_Fare)
plt.title('Fare distibution showing outliers')

plt.subplot(1,2,2)
sns.boxplot(y = df['Fare'] , width = .4)
plt.title('Distribution or Fare')
plt.tight_layout();
```

```python
# most peaple died in the titanic
sns.histplot(x=df['Survived']);
plt.title('Count of people Survived and not');
```

```python
# most passengers were males
sns.countplot(x= df['Sex']);
```

```python
# females were prioritized to survive and males sacrificed them selvse 
sns.countplot(x = df['Sex'] , hue = df['Survived'])
plt.title('Survived numbers among genders');
```

```python
# the high classes have more survival rate
sns.countplot(x = df['Pclass'] , hue = df['Survived'])
plt.title('Survive distributin among classes');
```

```python
# females where prioritized in all classes to survive and also the higher class has higher survival rate
sns.catplot(x = df['Pclass'] , y = df['Survived'], hue = df['Sex'] , kind = 'point')
plt.title('Survival distribution among the classes and genders');
```

```python
# most passengers were from Southampton embark
sns.countplot(x = df['Embarked']);
```

```python
# Cherbourg only has more survivors than its non-survivors but there is no obvious relation between the survival and embarkation
sns.countplot(x = df['Embarked'] , hue = df['Survived'])
plt.title('Survival distribution among embarkation');
```

```python
# children below 8 years were prioritized
sns.histplot(x = df['Age'] , hue = df['Survived'] , bins = 20)
plt.title('Survival distribution among ages');
```

```python
# being in family did not show a true increase in survival rate
plt.subplot(1,2,1)
sns.countplot(x = df['Parch'], hue = df['Survived'])
plt.xlabel('Number of parents and children for a passenger with him')
plt.subplot(1,2,2)
sns.countplot(x = df['SibSp'], hue = df['Survived'])
plt.xlabel('Number of siblings and spouses for a passenger with him')
plt.tight_layout();
```

```python
# the first class (1) got the highest fare
sns.boxplot(x = df['Pclass'] , y = df['Fare'])
plt.xlabel('Classes')
plt.title('Distribution of Fare among classes');
```

```python
# most people were in the 3rd class (lowest class)
sns.countplot(x = df['Pclass']);
```

```python
# the peaple in the high class are likely to be older
sns.boxplot(y= df['Age'] , x = df['Pclass'])
plt.title('Age Distribution among classes')
plt.xlabel('Classes');
```
